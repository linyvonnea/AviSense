# Model Training and Cross-Validation

This notebook summarizes the cross-validation results of the classical machine learning models used for Philippine bird species classification.

```python
from pathlib import Path
import pandas as pd
from IPython.display import Image, display

PROJECT_ROOT = Path("..")
REPORTS_DIR = PROJECT_ROOT / "outputs" / "reports"
FIGURES_DIR = PROJECT_ROOT / "outputs" / "figures"
```

## Cross-Validation Results

The models compared include Extra Trees, Random Forest, Logistic Regression, XGBoost, SVM with RBF kernel, Gradient Boosting, and K-Nearest Neighbors.

```python
possible_files = [
    REPORTS_DIR / "model_comparison.csv",
    REPORTS_DIR / "cv_results.csv",
    REPORTS_DIR / "cross_validation_results.csv"
]

for file in possible_files:
    print(file, file.exists())
```

```python
model_comparison_path = REPORTS_DIR / "model_comparison.csv"

if model_comparison_path.exists():
    model_comparison = pd.read_csv(model_comparison_path)
    display(model_comparison)
else:
    print("model_comparison.csv not found. Check the exact filename in outputs/reports.")
```

```python
model_comparison_fig = FIGURES_DIR / "model_comparison.png"

if model_comparison_fig.exists():
    display(Image(filename=str(model_comparison_fig)))
else:
    print("model_comparison.png not found.")
```

## Cross-Validation Summary

Based on the cross-validation results, tree-based ensemble models performed strongly. Random Forest achieved the highest mean accuracy, while Extra Trees achieved the highest mean macro F1-score among the baseline models. XGBoost, Logistic Regression, and SVM also produced competitive results.
