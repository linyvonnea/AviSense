# Hyperparameter Tuning

This notebook summarizes the hyperparameter tuning results for the best-performing classical machine learning models.

```python
from pathlib import Path
import pandas as pd

PROJECT_ROOT = Path("..")
REPORTS_DIR = PROJECT_ROOT / "outputs" / "reports"
```

```python
for file in REPORTS_DIR.glob("*tuned*"):
    print(file)
```

```python
possible_tuning_files = [
    REPORTS_DIR / "tuning_results.csv",
    REPORTS_DIR / "tuned_results.csv",
    REPORTS_DIR / "best_params.csv"
]

for file in possible_tuning_files:
    print(file, file.exists())
```

```python
tuning_results = pd.DataFrame({
    "model": [
        "xgboost_tuned",
        "extra_trees_tuned",
        "random_forest_tuned",
        "svm_rbf_tuned"
    ],
    "best_cv_f1_macro": [
        0.771417605004183,
        0.7622232799868008,
        0.7536844498162176,
        0.7502782054851086
    ],
    "best_params": [
        "{'colsample_bytree': 0.9, 'learning_rate': 0.1, 'max_depth': 4, 'n_estimators': 100, 'subsample': 0.8}",
        "{'max_depth': None, 'max_features': 'sqrt', 'min_samples_leaf': 2, 'n_estimators': 300}",
        "{'max_depth': None, 'max_features': 'sqrt', 'min_samples_leaf': 1, 'n_estimators': 700}",
        "{'model__C': 50, 'model__gamma': 0.001}"
    ]
})

display(tuning_results)
```

## Tuning Summary

Among the tuned models, XGBoost obtained the highest cross-validation macro F1-score of 0.7714. The selected parameters were colsample_bytree = 0.9, learning_rate = 0.1, max_depth = 4, n_estimators = 100, and subsample = 0.8.
